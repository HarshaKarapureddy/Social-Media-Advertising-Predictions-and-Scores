```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
clean_df = pd.read_csv("campaigns_clean.csv")
```

```python
clean_df["CTR"] = clean_df["Clicks"] / clean_df["Impressions"]
```

```python
perf_cols = ["Impressions", "Engagement_Score", "Conversion_Rate", "CTR", "ROI"]
clean_df[perf_cols].describe()
```

CHANNEL CHOICE

```python
channel_perf = (
    clean_df
    .groupby("Channel_Used")[["ROI", "Conversion_Rate", "Engagement_Score", "CTR"]]
    .mean()
    .sort_values("ROI", ascending=False)
)

channel_perf
```

LANGUAGE CHOICE

```python
language_perf = (
    clean_df
    .groupby("Language")[["ROI", "Conversion_Rate", "Engagement_Score", "CTR"]]
    .mean()
    .sort_values("ROI", ascending=False)
)

language_perf
```

AUDIENCE CHOICE

```python
audience_perf = (
    clean_df
    .groupby("Target_Audience")[["ROI", "Conversion_Rate", "Engagement_Score", "CTR"]]
    .mean()
    .sort_values("ROI", ascending=False)
)

audience_perf
```

LOCATION CHOICE

```python
location_perf = (
    clean_df
    .groupby("Location")[["ROI", "Conversion_Rate", "Engagement_Score", "CTR"]]
    .mean()
    .sort_values("ROI", ascending=False)
)

location_perf
```

CAMPAIGN GOAL

```python
goal_perf = (
    clean_df
    .groupby("Campaign_Goal")[["ROI", "Conversion_Rate", "Engagement_Score", "CTR"]]
    .mean()
    .sort_values("ROI", ascending=False)
)

goal_perf
```

SEASON CHOICE

```python
season_perf = (
    clean_df
    .groupby("Season")[["ROI", "Conversion_Rate", "Engagement_Score", "CTR"]]
    .mean()
    .sort_values("ROI", ascending=False)
)

season_perf
```
