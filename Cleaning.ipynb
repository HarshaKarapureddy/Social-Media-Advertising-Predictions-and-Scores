```python
import pandas as pd

df = pd.read_csv("Social_Media_Advertising.csv")
```

```python
print("Shape:", df.shape)
display(df.head(5))
print("\nColumns:", df.columns.tolist())
df.info()
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
print(df.columns.tolist())
```

```python
print("Duplicates:", df.duplicated().sum())
```

```python
df["Duration_Days"] = (
    df["Duration"]
    .str.extract(r"(\d+)")
    .astype(int)
)

df[["Duration", "Duration_Days"]].head()
```

```python
df["Acquisition_Cost_Num"] = (
    df["Acquisition_Cost"]
    .replace(r"[\$,]", "", regex=True)
    .astype(float)
)

df[["Acquisition_Cost", "Acquisition_Cost_Num"]].head()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
df["Date"].head()
```

```python
df["Date"].isna().sum()
```

```python
df["Month"] = df["Date"].dt.month
df[["Date", "Month"]].head()
```

```python
df["Season"] = df["Month"].map({
    12:"Winter", 1:"Winter", 2:"Winter",
    3:"Spring", 4:"Spring", 5:"Spring",
    6:"Summer", 7:"Summer", 8:"Summer",
    9:"Fall", 10:"Fall", 11:"Fall"
})

df[["Date", "Month", "Season"]].head()
```

```python
num_check_cols = [
    "Duration_Days", "Acquisition_Cost_Num",
    "Clicks", "Impressions", "Engagement_Score",
    "Conversion_Rate", "ROI"
]

display(df[num_check_cols].describe())
```

```python
checks = {
    "Clicks < 0": (df["Clicks"] < 0).sum(),
    "Impressions <= 0": (df["Impressions"] <= 0).sum(),
    "Conversion_Rate outside [0,1]": (~df["Conversion_Rate"].between(0, 1)).sum(),
    "ROI < 0": (df["ROI"] < 0).sum(),
    "Duration_Days <= 0": (df["Duration_Days"] <= 0).sum(),
    "Acquisition_Cost_Num < 0": (df["Acquisition_Cost_Num"] < 0).sum(),
    "Clicks > Impressions": (df["Clicks"] > df["Impressions"]).sum(),
}

checks
```

```python
cat_cols = [
    "Target_Audience",
    "Campaign_Goal",
    "Channel_Used",
    "Location",
    "Language",
    "Customer_Segment",
    "Company"
]

for c in cat_cols:
    df[c] = (
        df[c]
        .astype(str)
        .str.strip()
        .str.title()
    )
```

```python
for c in cat_cols:
    print(c, "→", df[c].nunique())
```

```python
df = df.drop(columns=["Duration", "Acquisition_Cost"])
```

```python
clean_df = df.copy()
```

```python
clean_df.to_csv("campaigns_clean.csv", index=False)
```
